```python
from NetworkSim.simulation.tools.distribution import Distribution
```

```python
d = Distribution(seed=0)
interarrival1 = [d.poisson() for i in range(1000000)]
interarrival2 = [d.pareto() for i in range(1000000)]
```

```python
import numpy as np
print(np.sum(interarrival1))
print(np.average(interarrival1))
print(np.sum(interarrival2))
print(np.average(interarrival2))
```

```python
import matplotlib.pyplot as plt
plt.plot(interarrival1)
```

```python
plt.plot(interarrival2)
```

```python
plt.hist(interarrival1, bins=100, density=True)
```

```python
plt.hist(interarrival2, bins=100, density=True)
```
